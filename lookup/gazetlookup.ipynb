```python
fname="data/capplacefolia.txt"
out="data/text"
corp=open(fname).readlines()
outf=open(out,"w")
for l in corp:
    if len(l)>1:
        outf.write(l.split()[0]+"\n")
    else:
        outf.write("\n")
outf.close()
```

```python
def inds_to_lines(indf,corp,taggedcorp):
    ind1=open(indf).readlines()
    corp1=open(corp).read()
    inds=[int(x.split()[0]) for x in ind1]
    charindex=0
    linecount=0
    lines=[]
    for i in inds:
        a=corp1[charindex:i].count("\n")
        linecount+=a
        lines.append(linecount)
        charindex=(i+1)
    return lines
```

```python
lines=inds_to_lines("out1","data/text","b")
```

```python
def taglines(corp,lines,out):
    corp1=open(corp).readlines()
    out1=open(out,"w")
    for i in range(len(corp1)):
        if i in lines:
            out1.write(corp1[i][:-1]+" LOC\n")
        else:
            out1.write(corp1[i])
```
